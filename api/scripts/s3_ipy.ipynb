```python
def extract_date(filename):
    """
    Extracts the date from the filename using regex.
    
    Args:
        filename (str): The name of the file.
        
    Returns:
        str: The extracted date in 'YYYY-MM-DD' format, or None if not found.
    """
    match = re.search(r'news_week_(\d{4}-\d{2}-\d{2})', filename)
    return match.group(1) if match else None
```

```python
import pandas as pd
import math

def split_csv_into_parts(file_path, parts=5):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)
    
    # Calculate the number of rows per part
    total_rows = len(df)
    chunk_size = math.ceil(total_rows / parts)
    
    # Split the DataFrame into parts
    split_dfs = [df[i:i+chunk_size] for i in range(0, total_rows, chunk_size)]
    
    return split_dfs

file_path = "news_week_2025-05-12 00_00_00.csv"  # Replace with your actual file path

split_parts = split_csv_into_parts(file_path)
for i, part in enumerate(split_parts):
    part.to_csv(f"part_{i+1}.csv", index=False)
```
