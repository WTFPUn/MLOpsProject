```python
from pathlib import Path
from FlagEmbedding import BGEM3FlagModel
import hdbscan
import umap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# ตอนนี้อ่าน Doc จาก txt ไปก่อน
txt_files = list(Path("/content").glob("*.txt"))

sentences = []
for txt_file in txt_files:
    with txt_file.open("r", encoding="utf-8") as f:
        full_text = f.read().strip()
        if full_text:
            sentences.append(full_text)

print(f"รวมได้ {len(sentences)} ไฟล์ข้อความ")

outputs = model.encode(sentences, return_dense=False, return_sparse=False, return_colbert_vecs=True)
colbert_vecs = outputs['colbert_vecs']

```

```python
import umap
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import OPTICS

# Pooled vectors from colbert_vecs
pooled_vecs = np.array([vec.mean(axis=0) for vec in colbert_vecs])


# ✅ Standardize
pooled_vecs = StandardScaler().fit_transform(pooled_vecs)

# ✅ UMAP → 2D
umap_2d = umap.UMAP(n_neighbors=2, min_dist=0.1, metric='cosine', random_state=42).fit_transform(pooled_vecs)

# ✅ OPTICS Clustering
optics = OPTICS(min_samples=2, metric='euclidean', xi=0.05, min_cluster_size=2)
labels = optics.fit_predict(umap_2d)

# ✅ Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x=umap_2d[:, 0], y=umap_2d[:, 1], hue=labels, palette='Set2', s=120)
for i, sentence in enumerate(sentences):
    plt.text(umap_2d[i, 0]+0.01, umap_2d[i, 1], f"{i}", fontsize=10)
plt.title("UMAP + OPTICS Clustering")
plt.xlabel("UMAP-1")
plt.ylabel("UMAP-2")
plt.legend(title="Cluster")
plt.tight_layout()
plt.show()

# ✅ Table output
df = pd.DataFrame({
    'Sentence': sentences,
    'Cluster': labels
})
df
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import pandas as pd
from datetime import date
import os # เพื่อใช้ตรวจสอบว่าไฟล์มีอยู่หรือไม่

# --- Global Variables for LLM (to avoid reloading) ---
MODEL_ID = "scb10x/typhoon2.1-gemma3-4b"
LLM_PIPELINE = None
LLM_TOKENIZER = None

def initialize_llm_model_and_pipeline():
    """
    Initializes the LLM model and tokenizer if they haven't been already.
    Returns the summarizer pipeline and tokenizer.
    """
    global LLM_PIPELINE, LLM_TOKENIZER
    if LLM_PIPELINE is None:
        print("Initializing LLM model and pipeline (this may take a few minutes)...")
        LLM_TOKENIZER = AutoTokenizer.from_pretrained(MODEL_ID)
        model = AutoModelForCausalLM.from_pretrained(
            MODEL_ID,
            device_map="auto",
            torch_dtype=torch.bfloat16, # Use bfloat16 for better performance if supported
            # load_in_4bit=True, # Uncomment for 4-bit quantization to save memory
        )
        LLM_PIPELINE = pipeline(
            "text-generation",
            model=model,
            tokenizer=LLM_TOKENIZER,
            torch_dtype=torch.bfloat16,
            device_map="auto",
        )
        print("LLM model and pipeline initialized successfully.")
    return LLM_PIPELINE, LLM_TOKENIZER
```

```python
# --- Function to Summarize News for a Cluster ---
def summarize_news_cluster(cluster_text: str, llm_pipeline, tokenizer_llm) -> str:
    """
    Summarizes a given block of text (representing a news cluster) using the LLM.

    Args:
        cluster_text (str): The combined text of news articles in a cluster.
        llm_pipeline: The pre-initialized Hugging Face text-generation pipeline.
        tokenizer_llm: The pre-initialized tokenizer for the LLM.

    Returns:
        str: The summarized text, or an error message if summarization fails.
    """
    prompt_template = (
        "<|im_start|>system\nคุณเป็นผู้ช่วย AI ที่เชี่ยวชาญด้านการวิเคราะห์ สรุปใจความสำคัญของข่าว และเรียงลำดับเหตุการณ์ตามไทม์ไลน์<|im_end|>\n"
        "<|im_start|>user\n"
        "สำหรับกลุ่มข่าวต่อไปนี้ โปรดดำเนินการดังนี้:\n"
        "1.  **สรุปข่าว:** สรุปใจความสำคัญของกลุ่มข่าวนี้เป็นภาษาไทยอย่างกระชับ\n"
        "2.  **ไทม์ไลน์เหตุการณ์:** เรียงลำดับเหตุการณ์ของข่าวในกลุ่มนี้ให้ชัดเจนที่สุดเท่าที่จะทำได้ หากมีวันที่และเวลาปรากฏในข่าว โปรดระบุด้วยในการเรียงไทม์ไลน์\n\n"
        "--- ข้อมูลข่าวกลุ่มนี้ ---\n"
        f"{cluster_text}\n"
        "--- สิ้นสุดข้อมูลข่าวกลุ่มนี้ ---\n\n"
        "กรุณาตอบตามรูปแบบ:\n"
        "**สรุปข่าว:**\n[บทสรุปข่าวของกลุ่มนี้]\n\n"
        "**ไทม์ไลน์เหตุการณ์:**\n[ไทม์ไลน์ของเหตุการณ์ในกลุ่มนี้]\n<|im_end|>\n"
        "<|im_start|>assistant\n"
        "**สรุปข่าว:**\n"
    )

    try:
        sequences = llm_pipeline(
            prompt_template,
            max_new_tokens=1200, # Adjust as needed
            do_sample=True,
            temperature=0.6,
            top_p=0.9,
            top_k=50,
            eos_token_id=tokenizer_llm.eos_token_id,
            pad_token_id=tokenizer_llm.pad_token_id if tokenizer_llm.pad_token_id is not None else tokenizer_llm.eos_token_id
        )

        generated_summary = ""
        if sequences and sequences[0]['generated_text']:
            full_output = sequences[0]['generated_text']
            assistant_marker = "<|im_start|>assistant\n"
            start_idx = full_output.find(assistant_marker)

            if start_idx != -1:
                # The prompt already primes the assistant to start with "**สรุปข่าว:**"
                generated_summary = full_output[start_idx + len(assistant_marker):].strip()
                if generated_summary.endswith("<|im_end|>"):
                    generated_summary = generated_summary[:-len("<|im_end|>")].strip()
            else:
                generated_summary = "ERROR: Could not parse assistant response. Full output: " + full_output
            return generated_summary
        else:
            return "ERROR: LLM did not return any sequences."

    except Exception as e:
        return f"EXCEPTION during LLM summarization: {str(e)}"
```

```python
# --- Function to Export/Append Summary to CSV ---
def export_summary_to_csv(run_date_str: str, summarized_news_content: str, csv_filename: str = "llm_summaries_archive.csv"):
    """
    Appends a new summary record (date, summarized_news) to a CSV file.
    Creates the file with headers if it doesn't exist.

    Args:
        run_date_str (str): The date of the summarization run (e.g., "YYYY-MM-DD").
        summarized_news_content (str): The summarized news text from the LLM.
        csv_filename (str): The name of the CSV file to append to or create.
    """
    data_to_append = {
        'date': [run_date_str],
        'summarized_news': [summarized_news_content]
    }
    df_new_record = pd.DataFrame(data_to_append)

    # Define the columns for the CSV
    csv_columns = ['date', 'summarized_news']

    file_exists = os.path.exists(csv_filename)

    try:
        if not file_exists:
            # File doesn't exist, create it with header
            df_new_record.to_csv(csv_filename, index=False, encoding='utf-8-sig', columns=csv_columns)
            print(f"Created new CSV file: {csv_filename} and added the first record.")
        else:
            # File exists, append without header
            df_new_record.to_csv(csv_filename, mode='a', header=False, index=False, encoding='utf-8-sig', columns=csv_columns)
            # print(f"Appended record to {csv_filename}") # Optional: print confirmation for each append
    except Exception as e:
        print(f"Error exporting to CSV '{csv_filename}': {e}")
```
